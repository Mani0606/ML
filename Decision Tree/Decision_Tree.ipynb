```python
import pandas as ps
from sklearn import tree
from  sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = ps.read_csv("titanic.csv")
med=df.Age.mean()
df['Age']=df.Age.fillna(med)
ds = df.drop(['Name','PassengerId','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
ds.head(31)
```

```python
le_Sex = LabelEncoder()
ds['n_Sex'] = le_Sex.fit_transform(ds['Sex'])
ds.head()
```

```python
final = ds.drop('Sex',axis='columns')
final.head()
```

```python
x = final.drop(['Survived'],axis='columns')
y = final['Survived']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
reg = tree.DecisionTreeClassifier()
reg.fit(x_train,y_train)
reg.predict(x_test)
```

```python
reg.score(x_test,y_test)
```
